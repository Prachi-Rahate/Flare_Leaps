```python
time = np.arange(0,1,0.001)
#print(time)
signal = 3*np.sin(2*np.pi*70*time) + np.sin(80*2.0*np.pi*time) + 5*np.sin(10*2.0*np.pi*time)
#print(signal)
```

```python
plt.plot(time,signal)
plt.show()
```

```python
ft = scipy.fftpack.fft(signal)
print(len(ft))
#bins = np.fft.fftfreq(signal.size,0.1)
bins = np.fft.fftfreq(1000)
#print(bins)
```

```python
plt.plot(bins//0.001,np.abs(ft))
plt.xlim(-5,100)
#plt.ylim(0,10)
plt.show()
```

```python
from astropy.timeseries import LombScargle
freq,power = LombScargle(bins,signal).autopower
plt.plot(freq,power)
plt.show()
```

# Example

```python
#https://docs.astropy.org/en/stable/api/astropy.timeseries.LombScargle.html

rand = np.random.default_rng(42)

t = 100 * rand.random(100)
#print (len(t)) #100
#print(t) #observations/values

y = np.sin(2 * np.pi * t) + rand.standard_normal(100)
#print(y)
#print(len(y)) #100

plt.scatter(t,y)
```

```python
frequency, power = LombScargle(t, y).autopower()
print(frequency)
print(power)

freq = frequency[np.argmax(power)] 
print(freq)

plt.plot(frequency,power)
plt.show()
```

# Lomb scargle

```python
t = np.arange(0,1,0.001)
#print(len(t))
y = 3*np.sin(2*np.pi*50*t) + np.sin(80*2.0*np.pi*t)
plt.plot(t,y)
```

```python
frequency, power = LombScargle(t, y).autopower()
#freq = frequency[np.argmax(power)]
#print(freq)
plt.xlim(0,100)
plt.plot(frequency,power)
plt.show()
```

```python
#discrete fourier transform

from scipy.fft import fftfreq

ft = scipy.fftpack.fft(y)
xf = fftfreq(1000, 0.001)   #fftfreq(no. of sample points = 1000, sampling space = 0.001)
plt.xlim(0,100)
plt.plot(xf,np.abs(ft))
plt.show()
```
