```python
import lightkurve as lk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
search_result = lk.search_lightcurve('TIC 1403938')
search_result[0]
```

```python
lc = search_result[0].download()
```

```python
lc.scatter()
#plt.savefig("flare_1403938")
```

```python
lc.plot()
plt.xlim(1518.75,1519)
#plt.savefig("flare_1403938_zoomed")
```

```python
lc.plot()
plt.xlim(1527,1528)
plt.ylim(500,650)
#plt.savefig("flare_1403938_zoomed2")
```

# remove outliers and nan values

```python
#remove outliers and nan values

lc = lc.remove_nans().remove_outliers()
lc.scatter()
```

```python
lc.to_fits(path='flare_TIC_1403938.fits', overwrite=True)
```
