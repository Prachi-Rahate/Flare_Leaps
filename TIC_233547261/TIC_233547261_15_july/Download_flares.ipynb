```python
import lightkurve as lk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
search_result = lk.search_lightcurve('TIC 233547261')
search_result[0]
```

```python
lc = search_result[0].download()
```

# remove outliers and nan values

```python
#remove outliers and nan values

lc1 = lc.remove_nans()#.remove_outliers()
#print(lc)
lc1 = (lc1["time", "flux", "flux_err"])
print(lc1)
```

```python
lc1.plot()
```

```python
#normalize

n_lc = lc1.normalize()
n_lc.plot()
```
